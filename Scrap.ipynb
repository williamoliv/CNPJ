```python
def unzip_all_files(folder_path):
    for item in os.listdir(folder_path):
        if item.endswith('.zip'):
            file_path = os.path.join(folder_path, item)
            with zipfile.ZipFile(file_path, 'r') as zip_ref:
                zip_ref.extractall(folder_path)
                print(f"Extracted {item}")

# Provide the path to the folder containing the zip files
folder_path = r'C:\Users\willi\Downloads\DataP'

# Call the function to unzip all files in the specified folder
unzip_all_files(folder_path)
```
